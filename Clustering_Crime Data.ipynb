## K Means 

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1=pd.read_csv('C:/Users/Acer/JUPYTER/Assignments/7.Clustring/crime_data.csv')
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
df_norm
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime3=pd.read_csv('C:/Users/Acer/JUPYTER/Assignments/7.Clustring/crime_data.csv')
crime3
```

```python
crime3.info()
```

```python
crime3.drop(['Unnamed: 0'],axis=1,inplace=True)
crime3
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime3)
crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
dbscan.labels_
```

```python
#cluster addition
crime3['clusters']=dbscan.labels_
crime3
```

```python
crime3.groupby('clusters').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime3['clusters'],crime3['UrbanPop'], c=dbscan.labels_) 
```
